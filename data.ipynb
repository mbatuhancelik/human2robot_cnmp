```python
import torch
import matplotlib.pyplot as plt
```

```python
landmark = torch.load("data/train/landmarks.pt")
```

```python
N_TRAJ = landmark.shape[0]
```

```python
landmark.shape
```

```python
fig, ax = plt.subplots(6, 6, figsize=(24, 24))
for i in range(6):
    for j in range(6):
        for k in range(N_TRAJ):
            idx = i * 6 + j
            if idx >= landmark.shape[2]:
                continue
            ax[i, j].plot(landmark[k, :, idx, 0], landmark[k, :, idx, 1], c="b", alpha=0.5)
```

```python
joint = torch.load("data/train/joints.pt")
```

```python
joint.shape
```

```python
fig, ax = plt.subplots(2, 4, figsize=(16, 8))
for i in range(2):
    for j in range(4):
        idx = i * 4 + j
        for k in range(N_TRAJ):
            if idx >= joint.shape[2]:
                continue
            ax[i, j].plot(joint[k, :, idx], c="r", alpha=0.5)
```

```python
position = torch.load("data/train/positions.pt")
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
for i in range(3):
    for k in range(N_TRAJ):
        ax[i].plot(position[k, :, i], c="g", alpha=0.5)
```
